```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# changing the default settings to display the data correctly
pd.set_option("display.max_rows", 1000)
pd.set_option("display.max_Columns", 500)
pd.set_option("display.width", 1000)

pollution = pd.read_csv("pollution.csv")
print(pollution)
```

```python
# Pair Plot
sns.set_style("whitegrid")
sns.pairplot(pollution, hue = "Air Quality", height = 3)
plt.show()
```

```python
# Distribution plot
plt.close()
sns.FacetGrid(pollution, hue = "Air Quality", height = 5) \
    .map(sns.distplot, "SO2 μg/l") \
    .add_legend()
plt.show()
```

```python
# Box Plot
plt.close()
sns.boxplot(x = "Air Quality", y = "SO2 μg/l", data = pollution)
plt.show()
```

```python
# Violin Plot
plt.close()
sns.violinplot(x = "Air Quality", y = "SO2 μg/l", data = pollution, size = 8)
plt.show()
```

```python
# Contour Plot
plt.close()
print(sns.jointplot(x = "SO2 μg/l", y = "PM10 μg/l", data = pollution, kind = "kde"))
plt.show()
```
